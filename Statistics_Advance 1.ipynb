Q.7 What is the significance of confidence intervals in statistical analysis?


  ->: A confidence interval (CI) gives a range of plausible values for a population parameter (e.g., mean) computed from sample data.

A 95% CI means: if we repeat the sampling procedure many times, approximately 95% of such intervals will contain the true parameter.

Significance: CIs quantify estimation uncertainty, provide more information than a point estimate, and help make decisions with quantified risk.

 Q.9 : Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.



   -->Expected output (typical):

Sample mean: close to 50 (e.g., 49.9 or 50.1)

Sample standard deviation: close to 5 (e.g., 4.95 to 5.05)
Histogram: bell-shaped centered ≈50.

Note: Because sampling is random, exact numbers vary each run. Setting np.random.seed(42) makes results reproducible.

```python
# Requires: numpy, matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate 1000 random numbers from N(50, 5^2)
mean_pop = 50
sd_pop = 5
sample = np.random.normal(loc=mean_pop, scale=sd_pop, size=1000)

# Compute sample mean and standard deviation
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation (unbiased)

print("Sample mean:", sample_mean)
print("Sample standard deviation:", sample_std)

# Draw histogram
plt.hist(sample, bins=30, edgecolor='black')
plt.title("Histogram of 1000 samples from N(50, 5^2)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(sample_mean, linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.legend()
plt.show()
```

```python
import numpy as np
import math

# Optional: use scipy if available for exact t critical
try:
    from scipy import stats
    use_scipy = True
except Exception:
    use_scipy = False

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

n = len(daily_sales)
sample_mean = np.mean(daily_sales)
sample_std = np.std(daily_sales, ddof=1)  # sample standard deviation
se = sample_std / math.sqrt(n)

alpha = 0.05
if use_scipy:
    t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
else:
    # approximate t_crit for df=19 at 95% CI (common value ~ 2.093)
    t_crit = 2.093  # fallback if scipy not available

margin = t_crit * se
ci_lower = sample_mean - margin
ci_upper = sample_mean + margin

print("n =", n)
print("Sample mean =", sample_mean)
print("Sample std (s) =", sample_std)
print(f"Standard error (SE) = {se:.4f}")
print(f"t critical (two-tailed, 95%) = {t_crit:.4f}")
print(f"95% CI for mean: [{ci_lower:.2f}, {ci_upper:.2f}]")
```
